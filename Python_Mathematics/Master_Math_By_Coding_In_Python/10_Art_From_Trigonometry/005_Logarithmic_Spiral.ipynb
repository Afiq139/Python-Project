# 005 Logarithmic Spiral

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,10*np.pi,1234)
k = -3

x = np.cos(t) * np.exp(t/k)
y = np.sin(t) * np.exp(t/k)

plt.plot(x,y,'r')
plt.axis('square')
plt.axis('off')

plt.show()
```
