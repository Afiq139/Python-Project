```python
import sympy as sym
from IPython.display import display
sym.init_printing()
```

```python
mu, alpha, sigma = sym.symbols('mu, alpha, sigma')

expr = sym.exp( (mu-alpha)**2 / (2*sigma**2))
display(expr)
```

```python
x,y = sym.symbols('x,y')
expr = x+4+2*y
display(expr.subs(x, 4))
expr.subs({x:-4, y:3})
```
