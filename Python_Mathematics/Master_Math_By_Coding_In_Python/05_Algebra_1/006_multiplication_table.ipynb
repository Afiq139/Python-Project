# 006 Exercise create a multiplication table

```python
import numpy as np
import sympy as sym 
from IPython.display import display, Math
```

```python
nums = range(1,11)

matmult = np.zeros((len(nums), len(nums)), dtype = int)

for rowi in nums:
    for coli in nums:
        matmult[rowi-1, coli-1] = rowi*coli

display(Math(sym.latex(sym.sympify(matmult))))
```
